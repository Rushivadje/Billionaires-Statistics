```python
Top_categories=BSD['category'].value_counts().head(10)
Top_categories
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=Top_categories.values,y=Top_categories.index,palette='viridis',width=0.6)
plt.xlabel('Number of Billionaires')
plt.ylabel('Category')
plt.title('Top Categories with the Highest Concentration of Billionaires')
plt.show()
```

```python
age_mean=BSD['age'].mean()
print(age_mean)
print('-----------------')
age_median=BSD['age'].median()
print(age_median)
```

```python
plt.figure(figsize=(12,6))
sns.histplot(BSD['age'].dropna(),bins=20,kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Billionaires')
plt.axvline(age_mean,color='red',linewidth=2,linestyle='dashed',label=f'Mean age:{age_mean:.2f}')
plt.axvline(age_median,color='green',linewidth=2,linestyle='dashed',label=f'Median age:{age_median:.2f}')
plt.legend()
plt.show()
print('Mean age:',age_mean)
print('Median age:',age_median)
```

```python
total_billionaires=len(BSD)
print('Total_billionaires:',total_billionaires)
```

```python
print(BSD['selfMade'].unique())
```

```python
billionairestype_counts=BSD['selfMade'].value_counts()
billionairestype_counts
```

```python
self_made_count=len(BSD[BSD['selfMade']==True])
print('self_made_count:',self_made_count)
print('---------------------')

inherited_count=len(BSD[BSD['selfMade']==False])
print('inherited_count:',inherited_count)
```

```python
self_made_proportion=(self_made_count/total_billionaires)*100
print(f'self_made_proportion is:{self_made_proportion:.2f}')
print('-----------------------------')

inherited_proportion=(inherited_count/total_billionaires)*100
print(f'inherited_proportion is:{inherited_proportion:.2f}')
```

```python
plt.figure(figsize=(12,6))
slices=[self_made_proportion,inherited_proportion]
labels=['self_made','inherited']
plt.pie(slices,labels=labels,startangle=90,explode=(0.01,0.05),shadow=True,autopct='%.2f%%',colors=['c','m'])
plt.title('Proportion of Self-Made vs Inherited Wealth')
plt.axis('equal')
plt.show()
```

```python
self_made=BSD[BSD['selfMade']==True]
inherited=BSD[BSD['selfMade']==False]
self_made_mean_networth=self_made['finalWorth'].mean()
inherited_mean_networth=inherited['finalWorth'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=['Self-Made', 'Inherited'], y=[self_made_mean_networth, inherited_mean_networth], palette='pastel')
plt.ylabel('Mean Net Worth (in billions)')
plt.title('Comparison of Mean Net Worth: Self-Made vs. Inherited Billionaires')
plt.show()
print(f"Mean Net Worth of Self-Made Billionaires: ${self_made_mean_networth:.2f} billion")
print(f"Mean Net Worth of Inherited Billionaires: ${inherited_mean_networth:.2f} billion")
```
